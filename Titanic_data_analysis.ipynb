```python
d1['Age']=d1['Age'].fillna(d1['Age'].median())
```

```python
d1['Embarked']=d1['Embarked'].fillna(d1['Embarked'].mode()[0])
```

```python
d1.isnull().sum() 
```

```python
d1['Sex']=d1['Sex'].map({'male':0,'female':1})
```

```python
d1['Embarked']=d1['Embarked'].map({'S':0,'C':1,'Q':2})
```

```python
sns.countplot(x='Survived',data=d1)
plt.title('survival count')
plt.show()
```

```python
sns.countplot(x='Survived',hue='Sex',data=d1)
plt.title('survival by gender')
plt.show()
```

```python
d1['Age'].hist(bins=30)
plt.xlim(0,80)
plt.xlabel('Age')
plt.ylabel('no of passengers')
plt.title('Age distribution')
plt.show()
```

```python
d1['Fare'].hist(bins=30)
plt.title('fare distribution')

plt.show()
```
